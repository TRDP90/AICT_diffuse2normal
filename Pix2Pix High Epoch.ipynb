Move Files

```
def move_files(source, destination): 
  # gather all files
  allfiles = os.listdir(source)
 
  # iterate on all files to move them to destination folder
  for f in allfiles:
    src_path = os.path.join(source, f)
    dst_path = os.path.join(destination, f)
    os.rename(src_path, dst_path)
```

```
move_files('/content/pytorch-CycleGAN-and-pix2pix/Textures_Split_Dataset/test/A','/content/pytorch-CycleGAN-and-pix2pix/A/test')
move_files('/content/pytorch-CycleGAN-and-pix2pix/Textures_Split_Dataset/val/A','/content/pytorch-CycleGAN-and-pix2pix/A/val' )
move_files('/content/pytorch-CycleGAN-and-pix2pix/Textures_Split_Dataset/train/A','/content/pytorch-CycleGAN-and-pix2pix/A/train')
move_files('/content/pytorch-CycleGAN-and-pix2pix/Textures_Split_Dataset/test/B','/content/pytorch-CycleGAN-and-pix2pix/B/test')
move_files('/content/pytorch-CycleGAN-and-pix2pix/Textures_Split_Dataset/val/B','/content/pytorch-CycleGAN-and-pix2pix/B/val')
move_files('/content/pytorch-CycleGAN-and-pix2pix/Textures_Split_Dataset/train/B','/content/pytorch-CycleGAN-and-pix2pix/B/train')
```

# Training

-   `python train.py --dataroot EUVP_Paired_Underwater_Scenes --name EUVP_pix2pix --model pix2pix`

Change the `--dataroot` and `--name` to your own dataset's path and model's name. Use `--batch_size` to change the batch size (the default is 1).  Add `--direction BtoA` if you want to train a model to transfrom from class B to A (the default is AtoB).

Use `--n_epochs` to change the number of epochs with the initial learning rate (default is 100).  Use `--n_epochs_decay` to change the number of epochs to linearly decay learning rate to zero (default is 100).  The total of the `--n_epochs` + `--n_epochs_decay` is the number of epochs your model will run for.
